```python
large_lr = 'm4m_cache/tests/best_perf_sofar.grouped.valIsMeta/meta_medium-sized_mammals/0/weight_stats.pkl'
vis_data = vis_file(large_lr, smooth_factor=10)

# print('--'*60)

fldrs = glob('m4m_cache/tests/best_perf_sofar.grouped.valIsMeta/meta_medium-sized_mammals/')
display_fldr(fldrs[0], inner_key='0')
loc = "m4m_cache/tests/best_perf_sofar.separated.valIsMeta/meta_medium-sized_mammals/0/metrics.pkl"
stats = pkl.load(open(loc, 'rb'))
print(vis_data.shape)
sm = [v[0] for v in stats['small_mammals']['train']]
mm = [v[0] for v in stats['medium-sized_mammals']['train']]
sm_grads = vis_data[0, :, 3]
mm_grads = vis_data[1, :, 3]

fig, ax = plt.subplots(1, 2, figsize=(15, 10))
ax[0].plot(sm, label='small_mammals loss')
ax[0].plot(mm, label='medium-sized_mammals loss')
ax[0].legend()

ax[1].plot(sm_grads, label='small_mammals gradNorm')
ax[1].plot(mm_grads, label='medium-sized_mammals gradNorm')
ax[1].legend()
plt.show()
```
